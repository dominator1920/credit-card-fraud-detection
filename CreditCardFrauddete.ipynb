2. Model Selection and Training

To divide the data for input and output (target) attributes.

```python
X=data1.drop(columns='Class',axis=1) #input
Y=data1['Class'] #output

x=X.values
y=Y.values
```

We will split the data into training and testing sets.

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from lazypredict.Supervised import LazyClassifier
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True,custom_metric=None)
models,predictions = clf.fit(x_train,x_test,y_train,y_test)
```

Now we will define the ML models and fit them on the training data individually:

```python
print(models)
```

a. Logistic Regression

```python
log_reg = LogisticRegression(max_iter=1000)

log_reg.fit(x_train,y_train)
```

b. Decision Tree Classifier

```python
classifier= DecisionTreeClassifier(criterion='entropy', random_state=0)  

classifier.fit(x_train, y_train)  
```

```python
c. Support Vector Machine
```

```python
svm1=SVC()
svm1.fit(x_train,y_train)
```

```python
3. Prediction and Evaluation
```

For each model, we will predict the result, obtain the accuracy score, plot the confusion matrix and get the classification report.

a. For Logistic Regression :

```python
predictions1=log_reg.predict(x_test)

print("Score :  ",log_reg.score(x_test,y_test))

skplt.metrics.plot_confusion_matrix(y_test,predictions1,normalize=True, title = 'Confusion Matrix for Logistic Regression')

print("\n\n",classification_report(y_test,predictions1),"\n")
```

```python
plt.plot(predictions1,y_test)
```

For Decision Tree Classifier :

```python
predictions2= classifier.predict(x_test)

print("Score:",classifier.score(x_test,y_test))

skplt.metrics.plot_confusion_matrix(y_test,predictions2, normalize=False, title = 'Confusion Matrix for Decision Tree Classifier')



print("\n\n",classification_report(y_test,predictions2),"\n")
```

```python
plt.plot(predictions2,y_test)
```

For SVM :

```python
predictions3= svm1.predict(x_test)

print("Score:",svm1.score(x_test,y_test))

skplt.metrics.plot_confusion_matrix(y_test,predictions3, normalize=False, title = 'Confusion Matrix for Support Vector Machine:')



print("\n\n",classification_report(y_test,predictions3),"\n")
```
